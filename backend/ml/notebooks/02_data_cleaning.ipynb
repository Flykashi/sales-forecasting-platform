```python
import pandas as pd
import numpy as np

pd.set_option("display.max_columns", None)
print("Libraries imported")
```

```python
train_df = pd.read_csv("data/raw/train.csv")
store_df = pd.read_csv("data/raw/store.csv")

print("Train shape:", train_df.shape)
print("Store shape:", store_df.shape)
```

```python
df = pd.merge(train_df, store_df, on="Store", how="left")
print("Merged shape:", df.shape)
df.head()
```

convert date column

```python
df["Date"] = pd.to_datetime(df["Date"])
print(df["Date"].dtype)
```

handling missing vals 

```python
missing_values = df.isnull().sum().sort_values(ascending=False)
missing_values[missing_values > 0]
```

```python
df = df[df["Open"] == 1]
df = df[df["Sales"] > 0]

print("Shape after removing closed stores and zero sales rows:", df.shape)
```

finding the numerice miss ing values 

```python
df["CompetitionDistance"] = df["CompetitionDistance"].fillna(df["CompetitionDistance"].median())

df["CompetitionOpenSinceMonth"] = df["CompetitionOpenSinceMonth"].fillna(0)
df["CompetitionOpenSinceYear"] = df["CompetitionOpenSinceYear"].fillna(0)
df["Promo2SinceWeek"] = df["Promo2SinceWeek"].fillna(0)
df["Promo2SinceYear"] = df["Promo2SinceYear"].fillna(0)
```

```python
df["PromoInterval"] = df["PromoInterval"].fillna("None")
df["StateHoliday"] = df["StateHoliday"].fillna("0")
```

```python
df.isnull().sum().sort_values(ascending=False).head(20)
```

duplicate check

```python
print("Duplicate rows:", df.duplicated().sum())
```

saving the cleaned data

```python
df.to_csv("data/processed/cleaned_train.csv", index=False)
print("Cleaned data saved to /data/processed/cleaned_train.csv")
```
