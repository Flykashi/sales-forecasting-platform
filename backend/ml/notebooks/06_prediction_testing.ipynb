```python
import pandas as pd
import joblib
import numpy as np

```

```python
model = joblib.load("artifacts/model.pkl")
columns = joblib.load("artifacts/columns.pkl")

print("Model and columns loaded")
print("Total columns:", len(columns))
```

```python
sample_input = {col: 0 for col in columns}
sample_input
```

```python
sample_input["Store"] = 1
sample_input["DayOfWeek"] = 2
sample_input["Promo"] = 1
sample_input["SchoolHoliday"] = 0
sample_input["CompetitionDistance"] = 500.0
sample_input["Promo2"] = 0
sample_input["Year"] = 2015
sample_input["Month"] = 7
sample_input["Day"] = 15
sample_input["WeekOfYear"] = 29
sample_input["DayOfWeekNum"] = 2
sample_input["IsWeekend"] = 0
sample_input["IsMonthStart"] = 0
sample_input["IsMonthEnd"] = 0
sample_input["CompetitionOpenYears"] = 5
sample_input["Promo2RunningYears"] = 0
```

```python
for col in ["StoreType_b", "StoreType_c", "StoreType_d",
            "Assortment_b", "Assortment_c",
            "StateHoliday_a", "StateHoliday_b", "StateHoliday_c"]:
    if col in sample_input:
        sample_input[col] = 0

if "StoreType_b" in sample_input:
    sample_input["StoreType_b"] = 1

if "Assortment_b" in sample_input:
    sample_input["Assortment_b"] = 1
```

```python
sample_df = pd.DataFrame([sample_input])
sample_df.head()
```

```python
sample_df = sample_df[columns]
print(sample_df.shape)
```

```python
prediction_log = model.predict(sample_df)[0]
prediction_actual = float(pd.Series([prediction_log]).apply(lambda x: np.expm1(x))[0])

print("Predicted Sales:", prediction_actual)
```

DIFFERENT TYPES OF ERRORS 

```python
actual_sales = 6500   # replace with real original value
predicted_sales = 6229.638631523439

error = actual_sales - predicted_sales
print("Error:", error)
```

```python
absolute_error = abs(actual_sales - predicted_sales)
print("Absolute Error:", absolute_error)
```

```python
percentage_error = (abs(actual_sales - predicted_sales) / actual_sales) * 100
print("Percentage Error:", percentage_error, "%")
```
